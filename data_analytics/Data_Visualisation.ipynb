# Introduction

Remote work is an employment arrangement in which employees do their jobs in a location outside of a traditional office environment. This practice has been growing steadily over the past years; however it wasn't until the outbreak of COVID-19 that many organizations abruptly shifted to remote work to offer it as an alternative to employees. For example, before COVID-19, around 17% of employees in US worked from home whereas by the end of 2021, 58.6% of them was working remotely (Anderson, 2022).

This research captures the experiences, perceived benefits and challenges of remote work from those who, pre pandemic, had very limited exposure or no experience at all with remote work. For this purpose, people from different industries, contries and age were surveyed (Nole, 2022).

## Background

Discussions and research about remote work challenges and benefits due to COVID-19 have been addressed mostly by industry studies.
Back in 2020, as the pandemic started, GitLab released a report about Remote Work and had as objective to determine if remote work could scale. Nowadays, the picture regarding remote work is clearer, hence its most recent release of 2021, GitLab shift the focus of its report and aims to understand the challenges of remote work and what’s next.
Amongst its findings, GitLab identified that while 68% of remote workers plan to continue working from home after the pandemic ends, 42% of respondents acknowledged they have problems to maintain and set boundaries between work & home (GitLab, 2021).
Regarding how people feel about remote work, in its “2022 State of Remote Work” report, Buffer discovered that a whopping 97% of respondents would recommend remote work to others. Moreover, when describing their experience with remote work, 61% of them rated it as very positive, 29% somewhat positive, 9% neutral, 1% somewhat negative and 0% very negative.
In terms of advantages and disadvantages of remote work, Buffer also found that having more control of how they spend their time and ability to work and live wherever they want to
reflect that flexibility in terms of location and working hours remain as the main benefit perceived by people, whereas unplugging and loneliness were the main challenges employees found in remote work.


## Research Objectives

The main aim of this study is to understand how people experienced remote work and to discover the advantages and challenges of it. The focus lies on people who, pre pandemic, have had very limited or no experience at all working remotely. To meet this aim, this research should:

According to industry experts, if an organization poorly manages remote work, it may potentially impact employee’s morale and productivity, whereas the opposite would not only boost productivity but also build employee loyalty (Fuscaldo, 2021). However, despite remote work widespread adoption and imminent new way-of-working in the long run, companies are still learning and adapting management practices for remote work.
It is in this context that it becomes fundamentally important not only to understand why employees prefer remote work, but also to identify what are the main challenges employees face, so that organizations can elaborate actions plans and remote work policies to improve remote work conditions, which will ultimately improve employees’ productivity, enhance collaboration, and turn an organization into an attractive employer (Nole, 2022).

****Word Frequency Analysis for Q4 (Benefits of Remote Work)****

```python
# import spacy and the english language model
import spacy
nlp = spacy.load("en_core_web_sm")
nlp.max_length = 1850000

# import library to count words
from collections import Counter

# analysis done first for anwers to Q4
# join all responses to Q4 one single large string
large_textb = df_benefits.Q4.str.cat(sep = ' ')

# create a spacy document as this will split large text into single words and tag them a 'verbs', 'nouns',etc.
sp_docb = nlp(large_textb, disable = ['ner'])

# do overall word frequency analysis to evaluate the most common words in large text.
# I indicate the lemma attribute so that variations of a word are mapped to one single word (e.g. walked and walking are forms of word walk)
text_wordsb = [token.lemma_ for token in sp_docb if not token.is_stop and not token.is_punct]
text_wordsb_freq = Counter(text_wordsb)

# print most top 20 words with highest occurrance frequency among all answers given to Q4
text_wordsb_freq.most_common(20)

# previous code was re used and adapted from
# http://robertorocha.info/using-nlp-to-analyze-open-ended-responses-in-surveys/
# accessed on 10/09/2022
```

With regards Q4, after the word frequency analysis, once can get a sense of what surveyees mentioned more frequently as main benefits. For example they mentioned often words such as time, commute (probably less), money, well/ness. Let's dive deeper with pattern matching

****Pattern Matching for Q4 (Benefits of Remote Work)****

```python
# match a sequence of word types to get further insights
# for example, what are most common Adjective + Nouns present in answers given to Q4 - Benefits of remote work

# import library for word matchg
from spacy.matcher import Matcher

# define matcher
matcherb = Matcher(nlp.vocab)

# define pattern. In this case, I want the combination of Adjective + Noun
patternb = [{'POS':'ADJ'}, {'POS':'NOUN'}]
matcherb.add('ADJ_PHRASE', [patternb])

# run matcher across whole document containing answers
matchesb = matcherb(sp_docb, as_spans=True)

# create empty list of Adjective+Noun combination
phrasesb = []

for span in matchesb:
    # fill in list of Adj+Nouns
    phrasesb.append(span.text.lower())
    phraseb_freq = Counter(phrasesb)

# print top 20 phrases in list
phraseb_freq.most_common(20)

# previous code was re used and adapted from
# http://robertorocha.info/using-nlp-to-analyze-open-ended-responses-in-surveys/
# accessed on 10/09/2022
```

When analyzing actual phrases, one can get a better understanding of what are the main benefits surveyees find. Examples:
* More time
* Better work (either productivity or workf life balance)
* Less distractions
* More personal time (could be related to previously found 'More Time')

It seems 'more time' is one of the main benefits surveyees perceive. I'll use phrase matching to get a better understanding of what they refer when talking about 'more time'

****Phrase Matching for Q4 (Benefits of Remote Work)****

```python
# phrase matcher to find words which occur more often near to 'more time'

# import library to match words to phrases
from spacy.matcher import PhraseMatcher

more_time_collocb = []

# create matcher
# 'attr' below ensures all instances are converted to lower-case so the search is case-insensitive
matcherb1 = PhraseMatcher(nlp.vocab, attr = 'LOWER')

# define pattern and run the matcher across whole document
patternb1 = [nlp.make_doc('more time')]
matcherb1.add('moreTime', patternb1)
matchesb1 = matcherb1(sp_docb)

for match_id, start, end in matchesb1:
    span = sp_docb[start-10 : end+10]
    more_time_collocb.extend([token.lemma_.lower() for token in span if not token.is_stop and not token.is_punct])

Counter(more_time_collocb).most_common(20)

# previous code was re used and adapted from
# http://robertorocha.info/using-nlp-to-analyze-open-ended-responses-in-surveys/
# accessed on 10/09/2022
```

After evaluating which words are more often present around the phrase 'more time' in answers, I can see that among the top are:
* Commute: in the sense, they get more time as result of less commuting
* Life, Spend, Family, Balance: it seems the extra time surveyees get from remote work (potentially of less commuting) allows them to spend more time around their families, have a better work life balance, and get more involved in the activities they do in their private life.

```python
# import libraries needed for analysis
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk import PorterStemmer
import re
nltk.download('stopwords')
```

```python
# create list of stop words (e.g. the, a, etc)
STOPWORDS=stopwords.words("english")

def format_text(text):
    """perform some actions to text such as stemming, removing punctuation and stopwords
    Input arguments: 'text' whcih is raw text
    Output: cleaned text containing only relevant words"""
    # use Porter Stemmer to extract base form of words
    ps=PorterStemmer()

    # remove punctuation in text
    text_cleaned="".join([x for x in text if x not in string.punctuation])

    # remove white spaces
    text_cleaned=re.sub(' +', ' ', text_cleaned)

    # make sure all words are lowercase
    text_cleaned=text_cleaned.lower()

    # split into a list
    tokens=text_cleaned.split(" ")

    # take only words which aren't stop words
    tokens=[token for token in tokens if token not in STOPWORDS]

    # bring cleaned words together
    text_cleaned=" ".join([ps.stem(token) for token in tokens])

    # return cleaned text
    return text_cleaned


# previous code was re used and adapted from
# https://github.com/vaisakhnambiar/Web-scraping/blob/master/Amazon%20Reviews%20Analysis.ipynb
# accessed 10/09/2022
```

```python
# see how function format_text works in dataframe containing the answers to the open ended questions Q5

# copy a dataframe
df_challenges1 = df_challenges.copy()

# apply format_text function to dataframe and see how it cleans text
df_challenges1['cleaned_Q5']=df_challenges1['Q5'].apply(lambda x:format_text(x))

df_challenges1.head(5)
```

## Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
